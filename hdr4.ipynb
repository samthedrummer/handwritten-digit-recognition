```python
def plot_predictions(images, labels, predictions, num_images=10):
    plt.figure(figsize=(10, 10))
    for i in range(num_images):
        plt.subplot(5, 2, i + 1)
        plt.imshow(images[i].reshape(8, 8), cmap='gray')
        plt.title(f'True: {labels[i]}, Pred: {predictions[i]}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

plot_predictions(test_images, test_label, test_predictions)
```
